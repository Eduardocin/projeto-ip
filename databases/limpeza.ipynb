# Limpeza do Energia 

primeiro, iremos extrair todas as sheets do arquivo com excessão do sumário

```python
import pandas as pd

arquivos_nomes = []

for i in range(12): # 12 é o número de sheets
    new_df = pd.read_excel(io='./Dados EPE/Energia e Socioeconomia.xlsx', sheet_name=(i + 1), header=9) # todas as tabelas começam com 9 linhas de header
    new_df.drop(new_df.columns[len(new_df.columns)-1], axis=1, inplace=True)
    arquivos_nomes.append(new_df)



```

agora, vamos fazer a limpeza mais fina em cada tabela e vamos trabspô-la para que os anos sejam as linhas 

```python
new_header = arquivos_nomes[0].iloc[0] #grab the first row for the header
arquivos_nomes[0] = arquivos_nomes[0][1:] #take the data less the header row
arquivos_nomes[0].columns = new_header #set the header row as the df header

arquivos_nomes[7]
```
